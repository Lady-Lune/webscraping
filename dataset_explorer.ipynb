```python
import pandas as pd
```

```python
file_path = "newscraper/newscraper/all_cat_all_day.json"
df = pd.read_json(file_path)
df
```

```python
years = df["year"].unique() 
months = df["month"].unique()

for year in years:
    for month in months:
        count = df[(df["year"] == year) & (df["month"] == month)].shape[0]
        print(f"Year: {year}, Month: {month}, Count: {count}")
```
